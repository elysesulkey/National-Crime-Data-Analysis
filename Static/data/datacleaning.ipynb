```python
import pandas as pd
import numpy as np
```

```python
# import csv
raw_data = "Crime_Data_USA.csv"
raw_data_df = pd.read_csv(raw_data)

# drop rows with missing values
raw_data2_df = raw_data_df.dropna(how='any')
```

```python
# Convert data types
raw_data2_df["Year"] = raw_data2_df["Year"].astype(int)
```

```python
raw_data2_df
```

```python
# list of values of State column
States = raw_data2_df['State'].unique()
States
```

```python
# list of values of City column
Cities = raw_data2_df['City'].unique()
Cities
```

```python
raw_data2_df["State"].replace({"NORTH CAROLINA 5, 6, 8": "NORTH CAROLINA"}, inplace=True)
raw_data2_df["City"].replace({"BALTIMORE5": "BALTIMORE", "MINNEAPOLIS6": "MINNEAPOLIS", "ROCHESTER6": "ROCHESTER", "INDEPENDENCE6": "INDEPENDENCE", "LAS VEGAS METROPOLITAN POLICE DEPARTMENT": "LAS VEGAS", "FAYETTEVILLE5": "FAYETTEVILLE", "HIGH POINT5": "HIGH POINT", "OKLAHOMA CITY6": "OKLAHOMA CITY", "PORTLAND10": "PORTLAND", "METROPOLITAN NASHVILLE POLICE DEPARTMENT": "NASHVILLE", "AUSTIN6": "AUSTIN", "CORPUS CRISTI6": "CORPUS CRISTI", "LAREDO6": "LAREDO", "PASADENA6": "PASADENA", "RICHARDSON6": "RICHARDSON"}, inplace=True)
```

```python
cleaned_data = raw_data2_df.rename(columns = {"Aggravated assault": "Aggravated Assault"})
cleaned_data
```

```python
# sort by State
cleaned_data.sort_values(by=["State"], inplace=True)
cleaned_data
```

```python
#export cleaned dataframe
cleaned_data.to_csv("cleaned_crime_data.csv", index = False)
```
